```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def check_column(df, col_name):

    print(f'Resumen para columna {col_name}')
    print(df[col_name].describe())
    print('Cantidad registros nulos:', df[col_name].isna().sum())
    print()




```

# Import the train data

```python
df = pd.read_csv('data/train.csv')
df.head()
```

# Check for nulls

```python
for col in df.columns:
    check_column(df, col)
```

```python
columnas_mantener = ['Survived',
                     'Pclass', #1st = Upper, 2nd = Middle, 3rd = Lower
                     'Sex', 
                     'Age', 
                     'SibSp', # of siblings / spouses aboard the Titanic
                     'Parch', # # of parents / children aboard the Titanic	
                     'Fare',
                     'Embarked']
df = df[columnas_mantener].copy()
df
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Embarked', hue='Survived')
```

# Calcular probabilidad de supervivencia para distintas columnas

```python
df[['Survived', 'Embarked']].groupby('Embarked').agg(['sum', 'count']).reset_index(drop=True)
```

```python
df_g = df[['Survived', 'Embarked']].groupby('Embarked').agg(['sum', 'count']).droplevel(level=0, axis=1)
df_g['prob_surv'] = df_g['sum']/df_g['count']
df_g
```

```python
df_g = df[['Survived', 'Age']].groupby('Age').agg(['sum', 'count']).droplevel(level=0, axis=1)
df_g['prob_surv'] = df_g['sum']/df_g['count']
df_g
```
